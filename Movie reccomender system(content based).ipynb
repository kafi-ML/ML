```python
import ast
```

```python
def extract_genre_names(text):
    try:
        genres_list = ast.literal_eval(text)
        return [genre['name'] for genre in genres_list]
    except:
        return []

# Apply to the whole dataset
movies['genres'] = movies['genres'].apply(extract_genre_names)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(extract_genre_names)
```

```python
movies.head()
```

```python
def extract_top_cast(text):
    try:
        cast_list = ast.literal_eval(text)
        # Get the first 3 cast names
        return [member['name'] for member in cast_list[:3]]
    except:
        return []

# Apply it to the 'cast' column
movies['cast'] = movies['cast'].apply(extract_top_cast)
```

```python
def extract_director(text):
    try:
        crew_list = ast.literal_eval(text)
        # Filter for job == 'Director'
        directors = [member['name'] for member in crew_list if member['job'] == 'Director']
        return directors
    except:
        return []

# Apply to the dataset
movies['crew'] = movies['crew'].apply(extract_director)
```

```python
# Function to remove spaces in each element of a list
def remove_spaces_from_list(lst):
    return [x.replace(" ", "") for x in lst]

# Apply this function to the selected columns
movies['genres'] = movies['genres'].apply(remove_spaces_from_list)
movies['keywords'] = movies['keywords'].apply(remove_spaces_from_list)
movies['cast'] = movies['cast'].apply(remove_spaces_from_list)
movies['crew'] = movies['crew'].apply(remove_spaces_from_list)
```

```python
movies.head()
```

```python
# Combine all relevant columns into a new column 'tags' after converting list fields to strings
movies['tags'] = (
    movies['overview'].astype(str) + ' ' +
    movies['genres'].apply(lambda x: ' '.join(x)) + ' ' +
    movies['keywords'].apply(lambda x: ' '.join(x)) + ' ' +
    movies['cast'].apply(lambda x: ' '.join(x)) + ' ' +
    movies['crew'].apply(lambda x: ' '.join(x))
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Initialize CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

# 2. Fit and transform the 'tags' column
vectors = cv.fit_transform(movies['tags']).toarray()

# 3. Optional: Check feature names
print(cv.get_feature_names_out())

```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie_title):
    movie_index = movies[movies['title'] == movie_title].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(movies.iloc[i[0]].title)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies['tags']).toarray()

# Step 2: Compute cosine similarity matrix
similarity = cosine_similarity(vectors)

# Step 3: Create function to recommend and show similarity scores
def recommend_with_scores(movie_title):
    index = movies[movies['title'] == movie_title].index[0]
    distances = list(enumerate(similarity[index]))
    sorted_distances = sorted(distances, key=lambda x: x[1], reverse=True)[1:11]  # top 10 excluding itself

    recommended_titles = []
    similarity_scores = []

    for i in sorted_distances:
        title = movies.iloc[i[0]].title
        score = i[1]
        recommended_titles.append(title)
        similarity_scores.append(score)
        print(f"{title} ")
```

```python
recommend_with_scores("The Dark Knight")
```
